```python
import numpy as np
import Q1DSolver as Qsolv
import Q1DWFUGenerator as Qgen
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(-4.0,4.0,1024)
U = Qgen.Udouble(x, U1 = 15.0)
UU = Qgen.Udouble(x, U1 = 7.5)
E, Wf = Qgen.StaticSolver(x,U)
```

```python
for i in range(6):
    plt.plot(Wf[i])
plt.show()
```

```python
PositionWf = (-Wf[0]+Wf[1])/np.sqrt(2) + 1j*0
```

```python
plt.plot(x[1:-1],PositionWf)
plt.show()
```

```python
PositionWf = (-Wf[0]+Wf[1])/np.sqrt(2) + 1j*0
PositionWf2 = (-Wf[0]+Wf[1])/np.sqrt(2) + 1j*0
for i in range(50):
    Out  = Qsolv.TDSE_1D_Solver(PositionWf,UU[1:-1],x[1:-1],1e-5,5000)
    Out2 = Qsolv.TDSE_split_operator(PositionWf2,UU[1:-1],x[1:-1],1e-5,5000)
    PositionWf = Out
    PositionWf2 = Out2
    fig, axs = plt.subplots(ncols = 4, figsize=(24,5))
    axs[0].plot(x[1:-1],UU[1:-1])
    axs[0].plot(x[1:-1],U[1:-1])
    axs[1].plot(x[1:-1],np.real(Out))
    axs[1].plot(x[1:-1],np.real(Out2))
    axs[2].plot(x[1:-1],np.imag(Out))
    axs[2].plot(x[1:-1],np.imag(Out2))
    axs[3].plot(x[1:-1],np.power(np.abs(Out),2))
    axs[3].plot(x[1:-1],np.power(np.abs(Out2),2))
    plt.show()
    plt.clf()
```
